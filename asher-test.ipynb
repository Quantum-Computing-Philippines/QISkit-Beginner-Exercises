```python
def simulate(qc=qc):
  from qiskit.tools.visualization import plot_histogram
  backend_sim = Aer.get_backend('qasm_simulator')
  shots = 1000

  job_sim = execute(qc, backend=backend_sim, shots=shots)
  result_sim = job_sim.result()
  answer = result_sim.get_counts()
  return plot_histogram(answer)
```

````python
# ------------------------ Running the Code ----------------------- #

def simulate(qc=qc):
  from qiskit.tools.visualization import plot_histogram
  backend_sim = Aer.get_backend('qasm_simulator')
  shots = 1000

  job_sim = execute(qc, backend=backend_sim, shots=shots)
  result_sim = job_sim.result()
  answer = result_sim.get_counts()
  return plot_histogram(answer)
````

##Translating to a Quantum Circuit

Using the circuit diagram above, we can translate it to a quantum circuit by using the following rules:

1. For each input bit in the XOR gate, create a CNOT gate such that the input of the CNOT gate is the input bit and the target is the output of the XOR gate:

image.png

2.  For each AND gate, create a Control-Control-NOT gate (also known as Toffoli gate) such that the input bits of the Control-Control-NOT gate is the input bits of the AND gate and the target bit is the output bit of the AND gate.

image.png

Using the 2 rules above, translate the full-adder circuit to a quantum circuit.

````python
# ------------------------ Running the Code ----------------------- #
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.tools.visualization import plot_histogram

# Create Quantum registers
qINa = QuantumRegister(2,name="a")  #register a
qINb = QuantumRegister(2,name="b")  #register b
qINd = QuantumRegister(6,name="d0") #register d0
cIN0 = QuantumRegister(2,name="ci0") #carry
qINs = QuantumRegister(3,name="s") 
cOUT0 = QuantumRegister(1,name="co0") #carry out
# Create Classical Registers
c = ClassicalRegister(3, name="cout")

# Create Quantum Cicruit
qc = QuantumCircuit(qINa,qINb,cIN0, qINd,  qINs, cOUT0, c)




# Create a superposition for all qubits
#qc.h(qINa)
#qc.h(qINb)



qc.x(0)
qc.x(1)
qc.x(2)
qc.x(3)


#qc.h(qINd)
#qc.h(cIN0)
#qc.h(qINs)
#qc.h(cOUT0)
# --------------------------------------------------------------------------- #

# Computing d00
# XOR GATE
qc.cx(qINa[0],qINd[0])
qc.cx(qINb[0],qINd[0])
qc.barrier()

# Computing the output = s0
# XOR GATE
qc.cx(qINd[0],qINs[0])
qc.cx(cIN0[0],qINs[0])

# Measure the output = s0
qc.measure(qINs[0],c[0])
qc.barrier()

# Computing d01
# AND GATE
qc.ccx(cIN0[0],qINd[0],qINd[1])
qc.barrier()

# Computing d02
# AND GATE
qc.ccx(qINa[0],qINb[0],qINd[2])

# Computing Co0
# XOR Gate
qc.cx(qINd[1], qINd[2])
qc.cx(qINd[2], cOUT0[0]) # this will be the output carry to be added in second circuit
qc.barrier()




# -------------CARRYING THE OUTPUT TO ANOTHER ADDER------------------------------- #

# Computing d10
# XOR GATE
qc.cx(qINa[1],qINd[1])
qc.cx(qINb[1],qINd[1])
qc.barrier()

# Computing the output = s1
# XOR GATE
qc.cx(qINd[1],qINs[1])
qc.cx(cOUT0[0],qINs[1])

# Measure the output = s1
qc.measure(qINs[1],c[1])
qc.barrier()
qc.barrier()

# Computing d11
# AND GATE
qc.ccx(cOUT0[0],qINd[1],qINd[2])
qc.barrier()

# Computing d12
# AND GATE
qc.ccx(qINa[1],qINb[1],qINd[3])

# Computing the output = s2
# XOR Gate
qc.cx(qINd[2], qINs[2])
qc.cx(qINd[3], qINs[2]) # this will be the output carry to be added in second circuit

# Measure the output = s2
qc.measure(qINs[2],c[2])



# Draw Circuit
qc.draw(output='mpl')
````
